```python
import matplotlib.pyplot as plt
from transformers import GPT2LMHeadModel, GPT2Tokenizer, AdamW
import torch
import pandas as pd
import torch.nn.functional as F

n=1

# List to store perplexity scores for visualization
perplexity_scores = []
perplexity_scores_2 = []

# Simulate feedback data collection
feedback_data = {
    'user_id': [1, 2, 3],
    'feedback': ['formal', 'casual', 'casual'],
    'response_quality': [4, 5, 3]  # Ratings from 1 to 5
}
feedback_df = pd.DataFrame(feedback_data)

def Rl():
  from transformers import GPT2LMHeadModel, GPT2Tokenizer, AdamW
  import torch
  import pandas as pd
  import matplotlib.pyplot as plt
  import math

  # Initialize the model and tokenizer
  model_name = 'gpt2'
  model = GPT2LMHeadModel.from_pretrained(model_name)
  tokenizer = GPT2Tokenizer.from_pretrained(model_name)

  # Assign the pad_token to eos_token to avoid padding errors
  tokenizer.pad_token = tokenizer.eos_token

  # Set the model to training mode for fine-tuning and RL
  model.train()

  # Initialize the optimizer for the policy update
  optimizer = AdamW(model.parameters(), lr=5e-5)

  # Function to generate a response with controlled output length
  def generate_response(prompt, style='casual'):
      inputs = tokenizer(prompt, return_tensors='pt', padding=True, truncation=True)

      # Generate a response
      outputs = model.generate(
          inputs.input_ids,
          max_length=len(inputs.input_ids[0]) + 5,  # Allow up to 5 additional tokens
          num_return_sequences=1,
          pad_token_id=tokenizer.eos_token_id,
          top_k=50,
          top_p=0.92,
          temperature=0.7,
          repetition_penalty=1.2,
          do_sample=True
      )

      # Decode the response
      response = tokenizer.decode(outputs[0], skip_special_tokens=True)

      # Construct response by combining prompt with relevant continuation
      response_parts = response.split(prompt, 1)

      if len(response_parts) > 1:
          # Extract the part after the prompt
          continuation = response_parts[1].strip()
      else:
          continuation = ""

      # Limit to only one or two additional words
      response_words = continuation.split()
      limited_response = ' '.join(response_words[:2]) if response_words else ''  # Get only the first two words

      # Combine the original prompt with the limited response
      final_response = f"{prompt} {limited_response}".strip() if limited_response else prompt

      # Adjust response based on style
      if style == 'formal':
          final_response = final_response.replace("I'm", "I am").replace("can't", "cannot")

      return final_response  # Return the final response

  # Function to compute perplexity
  def calculate_perplexity(prompt):
      inputs = tokenizer(prompt, return_tensors='pt', padding=True, truncation=True)
      with torch.no_grad():
          outputs = model(inputs.input_ids, labels=inputs.input_ids)
          loss = outputs.loss
          perplexity = math.exp(loss.item())
      return perplexity

  # Simulate feedback data collection
  global feedback_df, feedback_data
  feedback_data = {
      'user_id': [1, 2, 3],
      'feedback': ['formal', 'casual', 'casual'],
      'response_quality': [4, 5, 3]  # Ratings from 1 to 5
  }
  feedback_df = pd.DataFrame(feedback_data)

  def update_reward_model(feedback_df):
      reward_model = feedback_df.groupby('feedback').mean()['response_quality']
      return reward_model

  reward_model = update_reward_model(feedback_df)

  def fine_tune_model_with_rl(prompt, feedback_score):
      inputs = tokenizer(prompt, return_tensors='pt', padding=True, truncation=True)
      outputs = model(inputs.input_ids, labels=inputs.input_ids)

      # Calculate reward
      reward = feedback_score / 5.0  # Normalize feedback score
      loss = outputs.loss * reward  # Scale the loss by the reward

      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

      print(f"Model fine-tuned with feedback score {feedback_score} (RL loss: {loss.item()})")

  def adjust_model_based_on_feedback(style, reward_model):
      print(f"Adjusting model for style '{style}' with reward {reward_model.get(style, 0)}")

  def user_interaction():
      user_input = input("Enter your message: ")
      style_preference = input("Enter preferred style (casual/formal): ")
      if style_preference == '0':
        return 0,0

      response = generate_response(user_input, style=style_preference)
      print(f"Bot Response: {response}")

      feedback = input("Rate the response quality from 1 to 5: ")
      feedback_score = int(feedback)

      global feedback_df
      feedback_df = pd.concat([feedback_df, pd.DataFrame([{
          'user_id': 4,  # Example user ID
          'feedback': style_preference,
          'response_quality': feedback_score
      }])], ignore_index=True)

      updated_reward_model = update_reward_model(feedback_df)
      fine_tune_model_with_rl(user_input, feedback_score)

      # Calculate and display perplexity
      perplexity_scores_2.append(calculate_perplexity(user_input))

      return updated_reward_model, style_preference

  reward_model, style_preference = user_interaction()
  adjust_model_based_on_feedback(style_preference, reward_model)

def supervised():

  # Initialize the model and tokenizer
  model_name = 'gpt2'
  model = GPT2LMHeadModel.from_pretrained(model_name)
  tokenizer = GPT2Tokenizer.from_pretrained(model_name)

  # Assign the pad_token to eos_token to avoid padding errors
  tokenizer.pad_token = tokenizer.eos_token

  # Set the model to training mode for fine-tuning
  model.train()

  # Initialize the optimizer
  optimizer = AdamW(model.parameters(), lr=5e-5)


  # Function to generate a response with enhanced control
  def generate_response(prompt, style='casual', temperature=0.7, top_k=50, top_p=0.95):
      """
      Generate a response based on user input and desired style.
      """
      # Tokenize the input with attention mask
      inputs = tokenizer.encode(prompt, return_tensors='pt', padding=True, truncation=True)
      attention_mask = torch.ones_like(inputs)

      # Generate a response with controlled randomness (temperature, top_k, top_p)
      outputs = model.generate(
          inputs,
          max_length=100,
          num_return_sequences=1,
          attention_mask=attention_mask,
          pad_token_id=tokenizer.eos_token_id,
          temperature=temperature,  # Controls randomness of predictions
          top_k=top_k,  # Limits to top-k predictions
          top_p=top_p  # Limits cumulative probability
      )

      # Decode the output
      response = tokenizer.decode(outputs[0], skip_special_tokens=True)

      # Adjust the response based on style
      if style == 'formal':
          response = response.replace("I'm", "I am").replace("can't", "cannot")

      return response

  # Function to compute perplexity
  def compute_perplexity(prompt):
      """
      Compute the perplexity of the model for a given prompt.
      """
      # Tokenize the input
      inputs = tokenizer(prompt, return_tensors='pt')
      input_ids = inputs['input_ids']

      with torch.no_grad():
          outputs = model(input_ids, labels=input_ids)
          loss = outputs.loss

      # Calculate perplexity
      perplexity = torch.exp(loss)
      return perplexity.item()

  global feedback_df, feedback_data
  # Simulate feedback data collection
  feedback_data = {
      'user_id': [1, 2, 3],
      'feedback': ['formal', 'casual', 'casual'],
      'response_quality': [4, 5, 3]  # Ratings from 1 to 5
  }
  feedback_df = pd.DataFrame(feedback_data)

  def update_reward_model(feedback_df):
      """
      Update reward model based on feedback data.
      """
      reward_model = feedback_df.groupby('feedback').mean()['response_quality']
      return reward_model

  reward_model = update_reward_model(feedback_df)

  def fine_tune_model_on_feedback(prompt, feedback_score):
      """
      Fine-tune the model based on user feedback.
      """
      # Tokenize the input
      inputs = tokenizer.encode(prompt, return_tensors='pt')

      # Forward pass and calculate loss
      outputs = model(inputs, labels=inputs)
      loss = outputs.loss

      # Adjust loss based on feedback score (penalize low feedback)
      adjusted_loss = loss * (6 - feedback_score)  # Higher score, lower loss

      # Backpropagation
      optimizer.zero_grad()
      adjusted_loss.backward()
      optimizer.step()

      print(f"Model fine-tuned with feedback score {feedback_score} (adjusted loss: {adjusted_loss.item()})")

  def adjust_model_based_on_feedback(style, reward_model):
      """
      Simulate adjustment of model parameters based on feedback.
      """
      print(f"Adjusting model for style '{style}' with reward {reward_model.get(style, 0)}")

  def user_interaction():
      """
      Handle user input, generate responses, and collect feedback.
      """
      user_input = input("Enter your message: ")
      style_preference = input("Enter preferred style (casual/formal): ")
      if style_preference == "0":
        return 0,0

      # Generate a response based on style preference
      response = generate_response(user_input, style=style_preference)
      print(f"Bot Response: {response}")

      # Compute perplexity score for the input
      perplexity = compute_perplexity(user_input)
      print(f"Perplexity score: {perplexity}")

      # Store perplexity score for visualization
      perplexity_scores.append(perplexity)

      # Collect feedback
      feedback = input("Rate the response quality from 1 to 5: ")
      feedback_score = int(feedback)
      feedback_data = {
          'user_id': 4,  # Example user ID
          'feedback': style_preference,
          'response_quality': feedback_score
      }

      # Update feedback dataframe and reward model
      global feedback_df
      feedback_df = pd.concat([feedback_df, pd.DataFrame([feedback_data])], ignore_index=True)
      updated_reward_model = update_reward_model(feedback_df)

      # Fine-tune the model based on feedback
      fine_tune_model_on_feedback(user_input, feedback_score)

      return updated_reward_model, style_preference


  reward_model, style_preference = user_interaction()
  adjust_model_based_on_feedback(style_preference, reward_model)

# Function to plot perplexity scores
def plot_perplexity():
    """
    Plot perplexity scores over time.
    """
    print(perplexity_scores, perplexity_scores_2)
    plt.plot([i+1 for i in range(len(perplexity_scores))], perplexity_scores, label='Supervised')
    plt.scatter([i+1 for i in range(len(perplexity_scores))], perplexity_scores)
    for index, ele in enumerate(perplexity_scores):
      plt.annotate(round(ele,2), (index+1, ele), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8)
    plt.plot([i+1 for i in range(len(perplexity_scores_2))], perplexity_scores_2, label='RL')
    plt.scatter([i+1 for i in range(len(perplexity_scores_2))], perplexity_scores_2)
    for index, ele in enumerate(perplexity_scores_2):
      plt.annotate(round(ele,2), (index+1, ele), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8)
    plt.xlabel('Interaction Number')
    plt.ylabel('Perplexity Score')
    plt.title('Perplexity Over Time')
    plt.legend()
    plt.show()
    global n
    plt.savefig(f"comparison_chart_{n}.png")
    n+=1

# Main interaction loop
if __name__ == "__main__":
  while True:
    print("\n","---"*20,"Supervised","---"*20)
    supervised()
    print("---"*20,"RL","---"*20)
    Rl()
    print("---"*20,"Perplexity Scores","---"*20)
    plot_perplexity()
```
